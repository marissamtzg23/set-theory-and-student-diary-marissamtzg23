**TEORIA DE CONJUNTOS**


La teoría de conjuntos es una rama fundamental de las matemáticas que estudia los conjuntos, que son colecciones de objetos o elementos. Estos objetos pueden ser cualquier cosa, como números, letras, personas, o incluso otros conjuntos. Los elementos dentro de un conjunto pueden ser finitos o infinitos.

**UNIVERSO**

El Universo es un término divertido que hace referencia al "conjunto universo" en teoría de conjuntos, que es el conjunto que contiene todos los elementos posibles en un contexto específico. Es una forma lúdica de hablar sobre el conjunto que lo abarca todo.
